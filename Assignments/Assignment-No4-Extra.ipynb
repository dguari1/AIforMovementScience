```python
# This file contains code to read each file in the gait folder and identify if the file corresponds to a
#person with injury or not

import pandas as pd
import os
import re
```

```python
#all the files 
files = os.listdir('gaitData/') #read all files in the folder
files = [f for f in files if 'txt' in f] #keep only txt files
demographicData = pd.read_csv('gaitData/RBDSinfo.csv', index_col=None)


for file in files:
    #read the file
    gaitData = pd.read_csv('./gaitData/'+file, sep='\t', index_col=None)
    subjectID = int(re.findall(r'\d+', file)[0]) #using redex to extract the number from the file name
    #find the subjectID in the dataframe and extract the 'Injury' information
    isInjury = False if demographicData[demographicData['SubjectID']==subjectID]['Injury'].item() == 'No' else True
```
