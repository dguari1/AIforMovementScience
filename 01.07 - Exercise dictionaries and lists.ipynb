````python
lowest_homework_average = min(homework_averages)
index_of_lowest_homework_average = homework_averages.index(lowest_homework_average)
name_of_lowest_homework_average = Students[dictionary_keys[index_of_lowest_homework_average]]['name']
print(f'Student with the lowest homework average is {name_of_lowest_homework_average}')
````

````python
high_homework_average = max(homework_averages)
index_of_high_homework_average = homework_averages.index(high_homework_average)
name_of_high_homework_average = Students[dictionary_keys[index_of_high_homework_average]]['name']
print(f'Student with the highest homework average is {name_of_high_homework_average}')
````
