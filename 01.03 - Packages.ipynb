## Installing packages
Installing packages using pip (Package Installer for Python) or conda is straightforward (most times:-P)

Some basic Python packages that provide everyday functionalities include:
- os
- shutil
- time
- datetime

Some packages that you will find in your data analysis journey are:
- numpy -- Operating and manipulating arrays
- pandas -- Operating and manipulating tables
- matplotlib -- Making graphics
- seaborn -- Making pretty graphics pr
- scikit-learn -- Machine learning 
- scipy -- Data analysis and statistics
- statsmodels -- Statistics 
- pytorch -- Deep learning 
- tensorFlow -- Deep learning

Installing a package can be achieved in several ways:

- Pip
```
!pip install <package_name>
```
- Conda
```
!conda install <package_name>
```

````python
#for example: 
!pip install numpy
````

Here I'm trying to install the package **numpy** but it is already available in my virtual environment. To upgrade a package, you can use
```
!pip install <package_name> --upgrade
```

```
!pip install <package_name>==version
```

There are thousands of great Python packages available. These packages are in ```GitHub``` or ```PyPI```. If you find one interesting, please share it in the forum
